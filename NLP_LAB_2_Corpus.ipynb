```python
ps = PorterStemmer()
lm = WordNetLemmatizer()
```

```python
def preprocess(text):
    text = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    ftext = [txt for txt in text if not txt.lower() in stop_words]
    stext = [ps.stem(txt) for txt in ftext]
    ltext = [lm.lemmatize(txt) for txt in stext]
    text = " ".join(ltext)
    print(text)
```

```python
preprocess(brown_sent)
```

```python
preprocess(gutenberg_sent)
```
